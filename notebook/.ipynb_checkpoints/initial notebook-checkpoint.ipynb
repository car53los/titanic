```python
#相関のヒートマップ作成
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(10, 8))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=10)

correlation_heatmap(train_df)
```

## データ加工

### データの分解、結合

```python
train_df_cp = train_df.copy()
test_df_cp = test_df.copy()

Y = train_df_cp['Survived']
train_df_cp = train_df_cp.drop('Survived',axis=1)

train_df_cp['cat'] = 'train'
test_df_cp['cat'] = 'test'
X_all = pd.concat([train_df_cp,test_df_cp],axis=0).reset_index(drop = True)
```

### initialの特徴の保存

```python
class ftr_init(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name]
        
for c in X_all.columns:
    ftr_init(c).run(c).save()
```

### Family Size、Aloneフラグの作成

```python
class FamilySize(ftr.Feature):
    def create_features(self,name=''):
        self.df['family_size'] = X_all['SibSp'] + X_all['Parch'] + 1
        
class Alone(ftr.Feature):
    def create_features(self,name=''):
        self.df['alone'] = X_all.apply(lambda x : 0 if x['SibSp'] + x['Parch'] ==0 else 1,axis=1)      

FamilySize('FamilySize').run().save()
Alone('Alone').run().save()
```

### 欠損値の保管

```python
# 中央値で保管
class fill_median(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].median()) 

# 平均値で保管
class fill_mean(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].mean())

# 最頻値で保管
class fill_mode(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].mode()[0])                

        
# 年齢、料金は中央値or平均値で、embarkは最頻値で補完        
for c in ['Age','Fare']:
    fill_median('{}_median'.format(c)).run(c).save()               
    fill_mean('{}_median'.format(c)).run(c).save()        

fill_mode('{}_mode'.format('Embarked')).run('Embarked').save()            
```

### ラベル付

```python
class Cabin(ftr.Feature):
    def create_features(self,name):
        self.df['Cabin'] = X_all['Cabin'].str[0]
        self.df['Cabin'] = self.df['Cabin'].map({'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'T':8,np.nan:0})

class Sex(ftr.Feature):
    def create_features(self,name):
        self.df['Sex'] = X_all['Sex'].map({'male':0,'female':1})       
        
Cabin('Cabin_label').run().save()        
Sex('Sex_label').run().save()        
```

### One-hot-vectorを作成

```python
class Ohe(ftr.Feature):
    def create_features(self,name):
        self.df = pd.get_dummies(ftr.load_datasets(['Embarked']))

Ohe('Embarked_Ohe').run().save()     
```

### ホールドアウト分割

```python
from sklearn.model_selection import train_test_split
# X1 = X[["Pclass", "Sex", "Age", "Fare"]]

X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.2,
                                                    random_state=0)

# 学習用データ、検証用データの分割の正常終了を確認する
display(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

### パイプラインで最適モデルを探す

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier,RidgeClassifier
from sklearn.ensemble import VotingClassifier,RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier,ExtraTreesClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
# pipeline setting
pipelines = {
     'sgd': Pipeline([('scl',StandardScaler()),
                      ('est',SGDClassifier())]),
     
     'ridge':Pipeline([('scl',StandardScaler()),
                       ('est',RidgeClassifier(random_state=0))]),

#      'rf': Pipeline([('scl',StandardScaler()),
#                      ('est',RandomForestClassifier(random_state=0))]),
     
     'gbr1': Pipeline([('scl',StandardScaler()),
                      ('est',GradientBoostingClassifier(random_state=0))]),

     'gbr2': Pipeline([('scl',StandardScaler()),
                      ('est',GradientBoostingClassifier(n_estimators=250,
                                                       random_state=0))]),
    
    'dt': Pipeline([('scl',StandardScaler()),
                      ('est',DecisionTreeClassifier())]),

    'ada': Pipeline([('est',AdaBoostClassifier())]),
    
    'bag': Pipeline([('est',BaggingClassifier())]),
    
    'et': Pipeline([('est',ExtraTreesClassifier())]),
    
    'gbr': Pipeline([('est',GradientBoostingClassifier())]),    

    'rf': Pipeline([('est',RandomForestClassifier())]),    
    
    'moderate':Pipeline([
        ('pca', PCA()),
        ('est',RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=25, max_features='auto', max_leaf_nodes=None,
            min_samples_leaf=1, min_samples_split=15,
            min_weight_fraction_leaf=0.0, n_estimators=51, n_jobs=4,
            oob_score=False, random_state=0, verbose=0, warm_start=False))]),
    
#     'moderate':Pipeline([
#         ('est',RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
#             max_depth=25, max_features='auto', max_leaf_nodes=None,
#             min_samples_leaf=1, min_samples_split=15,
#             min_weight_fraction_leaf=0.0, n_estimators=51, n_jobs=4,
#             oob_score=False, random_state=0, verbose=0, warm_start=False))])    
}

# build and evaluate
scores = {}
for pipe_name, pipeline in pipelines.items():
    pipeline.fit(X_train, Y_train)
    scores[(pipe_name,'1.train結果')] = accuracy_score(Y_train, pipeline.predict(X_train))
    scores[(pipe_name,'2.test結果')] = accuracy_score(Y_test, pipeline.predict(X_test))    
pd.Series(scores).unstack()    
```

### Voting

```python
vote_clf = VotingClassifier(estimators=[('rf', RandomForestClassifier()),
                                        ('gb',GradientBoostingClassifier()),
                                        ('et',ExtraTreesClassifier()),
                                        ('bag',BaggingClassifier()),
                                        ('ada',AdaBoostClassifier())
                                        ])
```

```python
vote_clf.fit(X_train, Y_train)
scores[('vote','1.train結果')] = accuracy_score(Y_train, vote_clf.predict(X_train))
scores[('vote','2.test結果')] = accuracy_score(Y_test, vote_clf.predict(X_test)) 
pd.Series(scores).unstack()    
```

### ニューラルネットワークモデルの検証

```python
import tensorflow as tf
from keras.models import Model, load_model
from keras.models import Sequential
from keras.layers import Input, Activation, Dense, LSTM, TimeDistributed, Lambda, Reshape, Permute, Multiply, Add, Concatenate
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score
import keras.backend as K
```

```python
# create model
model = Sequential()
model.add(Dense(64, input_shape=(len(X_train.columns),)))

for i in range(0, 8):
    model.add(Dense(units=64))
    model.add(Activation('relu'))

model.add(Dense(units=1))
model.add(Activation('linear'))

# model.compile(loss='mean_squared_error', optimizer='rmsprop')
model.compile(loss='categorical_crossentropy', metrics=["accuracy"], optimizer='adam')
model.fit(X_train, Y_train, epochs=50)
```

```python
accuracy_score(Y_test, model.predict_classes(X_test))
```
