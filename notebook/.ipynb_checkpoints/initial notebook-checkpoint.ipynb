## データ加工

### データの分解、結合

```python
train_df_cp = train_df.copy()
test_df_cp = test_df.copy()

Y = train_df_cp['Survived']
train_df_cp = train_df_cp.drop('Survived',axis=1)

train_df_cp['cat'] = 'train'
test_df_cp['cat'] = 'test'
X_all = pd.concat([train_df_cp,test_df_cp],axis=0).reset_index(drop = True)
```

### initialの特徴の保存

```python
class ftr_init(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name]

class ftr_init_answer(ftr.Feature):
    def create_features(self,name):
        self.df = pd.DataFrame(Y)        
        
for c in X_all.columns:
    ftr_init(c).run(c).save()

ftr_init_answer('Survived').run().save()
```

### Family Size、Aloneフラグの作成

```python
class FamilySize(ftr.Feature):
    def create_features(self,name=''):
        self.df['family_size'] = X_all['SibSp'] + X_all['Parch'] + 1
        
class Alone(ftr.Feature):
    def create_features(self,name=''):
        self.df['alone'] = X_all.apply(lambda x : 0 if x['SibSp'] + x['Parch'] ==0 else 1,axis=1)      

FamilySize('FamilySize').run().save()
Alone('Alone').run().save()
```

### 欠損値の保管

```python
# 中央値で保管
class fill_median(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].median()) 

# 平均値で保管
class fill_mean(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].mean())

# 最頻値で保管
class fill_mode(ftr.Feature):
    def create_features(self,name):
        self.df[name] = X_all[name].fillna(X_all[name].mode()[0])                

        
# 年齢、料金は中央値or平均値で、embarkは最頻値で補完        
for c in ['Age','Fare']:
    fill_median('{}_median'.format(c)).run(c).save()               
    fill_mean('{}_median'.format(c)).run(c).save()        

fill_mode('{}_mode'.format('Embarked')).run('Embarked').save()            
```

### ラベル付

```python
class Cabin(ftr.Feature):
    def create_features(self,name):
        self.df['Cabin'] = X_all['Cabin'].str[0]
        self.df['Cabin'] = self.df['Cabin'].map({'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'T':8,np.nan:0})

class Sex(ftr.Feature):
    def create_features(self,name):
        self.df['Sex'] = X_all['Sex'].map({'male':0,'female':1})       

class Embarked(ftr.Feature):
    def create_features(self,name):
        self.df['Embarked'] = ftr.load_datasets(['Embarked_mode'])['Embarked'].map({'S':0,'C':1,'Q':2})
        
Cabin('Cabin_label').run().save()        
Sex('Sex_label').run().save()        
Embarked('Embarked_label').run().save()  
```

### One-hot-vectorを作成

```python
class Ohe(ftr.Feature):
    def create_features(self,name):
        self.df = pd.get_dummies(ftr.load_datasets(['Embarked_mode']))

Ohe('Embarked_Ohe').run().save()     
```

### 名称タイトルを設定

```python
def replace_titles(x):
    title = x['Title']
    if title in ['Capt', 'Col', 'Don', 'Jonkheer', 'Major', 'Rev', 'Sir']:
        return 'Mr'
    elif title in ['the Countess', 'Mme', 'Lady','Mrs. Martin (Elizabeth L','Dona']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms']:
        return 'Miss'
    elif title =='Dr':
        if x['Sex']=='male':
            return 'Mr'
        else:
            return 'Mrs'
    else:
        return title
```

```python
class Title(ftr.Feature):
    def create_features(self,name):
        X_all['Title'] = X_all['Name'].str.extract(', (.+)\.')
        self.df['Title'] = X_all.apply(lambda x : replace_titles(x) ,axis =1)
        self.df['Title'] = self.df['Title'].map({'Master':0,'Mr':1,'Miss':2,'Mrs':3})
        
Title('Title').run().save() 
```

### パイプラインで最適モデルを探す

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier,RidgeClassifier
from sklearn.ensemble import VotingClassifier,RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier,ExtraTreesClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
import lightgbm as lgb

# pipeline setting
pipelines = {}

#sgd
model = {
    '1.sgd' : SGDClassifier(),
    '2.ridge' : RidgeClassifier(),
    '3.dt' : DecisionTreeClassifier(),
    '4.rf' : RandomForestClassifier(),
    '5.gbr' : GradientBoostingClassifier(),
    '6.ada' : AdaBoostClassifier(),
    '7.bag' : BaggingClassifier(),
    '8.et' : ExtraTreesClassifier(),
    '9.lgbm' : lgb.LGBMClassifier()
}
         
for m in model:
    pipelines[m] = Pipeline([('est',model[m])])
    pipelines['{}_pca'.format(m)] = Pipeline([('pca', PCA()),('est',model[m])])
    pipelines['{}_scl'.format(m)] = Pipeline([('scl',StandardScaler()),('est',model[m])])

#     'moderate2':Pipeline([
#         ('est',RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
#             max_depth=25, max_features='auto', max_leaf_nodes=None,
#             min_samples_leaf=1, min_samples_split=15,
#             min_weight_fraction_leaf=0.0, n_estimators=51, n_jobs=4,
#             oob_score=False, random_state=0, verbose=0, warm_start=False))])    
# }

# build and evaluate
scores = {}
for pipe_name, pipeline in pipelines.items():
    pipeline.fit(X_train, Y_train)
    scores[(pipe_name,'1.train結果')] = accuracy_score(Y_train, pipeline.predict(X_train))
    scores[(pipe_name,'2.test結果')] = accuracy_score(Y_test, pipeline.predict(X_test))    
pd.Series(scores).unstack()    
```

### Bayesian optを使用

```python
from hyperopt import hp, tpe, Trials, fmin
from sklearn.ensemble import VotingClassifier,RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier,ExtraTreesClassifier
from sklearn.metrics import accuracy_score

def objective(args):
    # モデルのインスタンス化
    rf = RandomForestClassifier(**args)
    # trainデータを使ってモデルの学習
    rf.fit(X_train, Y_train)
    # 今回はmicro f1を最大化したいので、-1をかけて最小化に合わせる
    return -1 * accuracy_score(Y_test, rf.predict(X_test))  
```

```python
hyperopt_parameters = {
#         'max_depth': hp.choice('max_depth', np.arange(10, 30, dtype=int)),
        'max_depth': hp.choice('max_depth', np.arange(10, 30, dtype=int)),    
#         'min_child_weight': hp.quniform ('min_child', 1, 20, 1),
#         'subsample': hp.uniform ('subsample', 0.8, 1),
        'n_estimators' : hp.choice('n_estimators', np.arange(1000, 10000, 100, dtype=int)),
#         'learning_rate' : hp.quniform('learning_rate', 0.025, 0.5, 0.025),
#         'gamma' : hp.quniform('gamma', 0.5, 1, 0.05),
#         'colsample_bytree' : hp.quniform('colsample_bytree', 0.5, 1, 0.05)
    }

# iterationする回数
max_evals = 10
# 試行の過程を記録するインスタンス
trials = Trials()

best = fmin(
    # 最小化する値を定義した関数
    objective,
    # 探索するパラメータのdictもしくはlist
    hyperopt_parameters,
    # どのロジックを利用するか、基本的にはtpe.suggestでok
    algo=tpe.suggest,
    max_evals=max_evals,
    trials=trials,
    # 試行の過程を出力
    verbose=1
)
```
